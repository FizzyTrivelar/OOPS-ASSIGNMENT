#Python OOPs Questions

1. What is Object-Oriented Programming (OOP)?  

Ans -  Object-Oriented Programming (OOP) is a programming paradigm that uses objects to represent data and methods. It organizes software design around data, or objects, rather than functions and logic. The four fundamental pillars of OOP are Inheritance, Polymorphism, Encapsulation, and Abstraction.

2. What is a class in OOP?  

Ans -  A class is a blueprint or prototype from which objects are created. It defines a set of attributes (data) and methods (functions) that the created objects of that class will possess.

3. What is an object in OOP?  

Ans -  An object is an instance of a class. When a class is defined, no memory is allocated until an object of that class is created. It contains actual values instead of blueprints.

4. What is the difference between abstraction and encapsulation?   

Ans -

a. Abstraction hiding internal implementation details and showing only the essential features to the outside world. (e.g., hiding how a car engine works, exposing only the steering wheel).

b. Encapsulation binding the data (attributes) and methods that manipulate that data together into a single unit (a class) and restricting direct access to some of the object's components.

5. What are dunder methods in Python?

Ans -   Dunder (Double Underscore) methods—also called magic methods—are special predefined methods in Python that start and end with double underscores (e.g., __init__, __str__). They allow you to emulate built-in type behaviors or implement operator overloading.

6. Explain the concept of inheritance in OOP.  

Ans -  Inheritance is a mechanism where a new class (child/derived class) acquires the attributes and methods of an existing class (parent/base class). This promotes code reusability.

7. What is polymorphism in OOP?

Ans -  Polymorphism means "many forms." It allows different classes to have methods with the identical name but different implementations, allowing an interface to be used for a general class of actions.

8. How is encapsulation achieved in Python?  

Ans - Encapsulation is achieved by using access modifiers:

Protected attributes: Prefixed with a single underscore _ (e.g., _variable).

Private attributes: Prefixed with a double underscore __ (e.g., __variable). Python uses name mangling internally to make these harder to access from outside the class.

9. What is a constructor in Python?  

Ans -  A constructor is a special method called automatically when an object of a class is instantiated. In Python, the __init__ method serves as the constructor and is typically used to initialize the object's state.

10. What are class and static methods in Python?  

Ans -

 a. Class Method: Bound to the class and not the object. It takes cls as its first parameter and can modify class state. Marked with @classmethod.

 b. Static Method: Bound to the class but doesn't have access to class (cls) or instance (self) states. It behaves like an isolated, regular function inside a class. Marked with @staticmethod.

 11. What is method overloading in Python?  

 Ans - Method overloading is the ability to define multiple methods with the same name but different signatures (parameters). Python does not support native method overloading. If you define multiple methods with the same name, the last one overrides the previous ones. However, it can be achieved using default arguments or variable-length arguments (*args, kwargs).

 12. What is method overriding in OOP?  

 Ans - Method overriding occurs when a child class provides a specific implementation for a method that is already defined in its parent class. The child's method overrides the parent's execution at runtime.

 13. What is a property decorator in Python?

 Ans -  The @property decorator allows you to define a method that can be accessed like a standard attribute, allowing you to implement getters, setters, and deleters safely while controlling data modification.

 14. Why is polymorphism important in OOP?   

 Ans - Polymorphism provides flexibility and scalability. It allows you to write clean, reusable, and generic code where a single function or method can process objects of completely different classes as long as they share a common interface.

 15. What is an abstract class in Python?

 Ans -   An abstract class is a class that cannot be instantiated directly and contains one or more abstract methods (methods declared without implementation). It acts as a template for other classes and is implemented using Python’s built-in abc module.

2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height


c = Circle(5)
r = Rectangle(4, 5)
print(f"Circle Area: {c.area():.2f}")
print(f"Rectangle Area: {r.area()}")
```

3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute.

```python
class Vehicle:
    def __init__(self, v_type):
        self.type = v_type

class Car(Vehicle):
    def __init__(self, v_type, brand):
        super().__init__(v_type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, v_type, brand, battery_capacity):
        super().__init__(v_type, brand)
        self.battery = battery_capacity

ev = ElectricCar("Sedan", "Tesla", "85 kWh")
print(f"Type: {ev.type}, Brand: {ev.brand}, Battery: {ev.battery}")
```

4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
Sparrow and Penguin that override the fly() method.


```python
class Bird:
    def fly(self):
        print("Most birds can fly.")

class Sparrow(Bird):
    def fly(self):
        print("Sparrows fly high in small flocks.")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly, they swim!")


birds = [Sparrow(), Penguin()]
for bird in birds:
    bird.fly()
```

5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance.

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. Current balance: {self.__balance}")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. Current balance: {self.__balance}")
        else:
            print("Insufficient balance or invalid amount.")

    def check_balance(self):
        return self.__balance


account = BankAccount(1000)
account.deposit(500)
account.withdraw(300)
print("Balance Check:", account.check_balance())
```

6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
and Piano that implement their own version of play().

```python
class Instrument:
    def play(self):
        print("Playing a generic instrument.")

class Guitar(Instrument):
    def play(self):
        print("Strumming guitar chords...")

class Piano(Instrument):
    def play(self):
        print("Playing a melodic piano piece...")

def start_performance(instrument_obj):
    instrument_obj.play()

start_performance(Guitar())
start_performance(Piano())
```

7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers.

```python
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print("Addition (Class Method):", MathOperations.add_numbers(10, 5))
print("Subtraction (Static Method):", MathOperations.subtract_numbers(10, 5))
```

10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 4)
v2 = Vector(1, 3)
v3 = v1 + v2
print("Summed Vector:", v3)
```

```python
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

s = Student("Mustafijur", [85, 90, 78, 92])
print(f"{s.name}'s Average Grade: {s.average_grade():.2f}")
```

14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary.

```python
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

emp = Employee("Amit", 160, 200)
mngr = Manager("Suresh", 160, 350, 15000)

print(f"Employee Salary: ₹{emp.calculate_salary()}")
print(f"Manager Salary (inc. bonus): ₹{mngr.calculate_salary()}")
```

16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

farm_animals = [Cow(), Sheep()]
for animal in farm_animals:
    animal.sound()
```

18. Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms.

```python
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

luxury_home = Mansion("Kanai Street", 4500000, 5)
print(f"Mansion Address: {luxury_home.address}")
print(f"Price: ₹{luxury_home.price}")
print(f"Rooms Available: {luxury_home.number_of_rooms}")
```
